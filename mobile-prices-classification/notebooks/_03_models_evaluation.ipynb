# Model Evaluation

```python
import pandas as pd
import numpy as np
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

## Prepairing models and tools

### Models

```python
lr = sklearn.linear_model.LogisticRegression()
nb = sklearn.naive_bayes.GaussianNB()
knn = sklearn.neighbors.KNeighborsClassifier()
svm = sklearn.svm.SVC()
# scaler
sc = sklearn.preprocessing.StandardScaler()
# cross validation
cv = sklearn.model_selection.cross_val_score
```

### Pipelines

to prevent data leakage during training and evaluation

```python
lr_pipeline = sklearn.pipeline.Pipeline(steps=[
    ('scaler', sc),
    ('lr', lr)
])
nb_pipeline = sklearn.pipeline.Pipeline(steps=[
    ('scaler', sc),
    ('nb', nb)
])
knn_pipeline = sklearn.pipeline.Pipeline(steps=[
    ('scaler', sc),
    ('knn', knn)
])
svm_pipeline = sklearn.pipeline.Pipeline(steps=[
    ('scaler', sc),
    ('svm', svm)
])
```

### datasets

```python
sf_x = pd.read_csv(r'C:\Users\Island\OneDrive\Desktop\ML Projects\ml-practice-lab\mobile-prices-classification\dataset\sf_x.csv')
sf_y = pd.read_csv(r'C:\Users\Island\OneDrive\Desktop\ML Projects\ml-practice-lab\mobile-prices-classification\dataset\sf_y.csv')
af_x = pd.read_csv(r'C:\Users\Island\OneDrive\Desktop\ML Projects\ml-practice-lab\mobile-prices-classification\dataset\af_x.csv')
af_y = pd.read_csv(r'C:\Users\Island\OneDrive\Desktop\ML Projects\ml-practice-lab\mobile-prices-classification\dataset\af_y.csv')
```

## evaluate the models

### Selected features dataset accuracies

```python
lr_sf_acc = cv(lr_pipeline, sf_x, sf_y, scoring='accuracy', cv=10)
nb_sf_acc = cv(nb_pipeline, sf_x, sf_y, scoring='accuracy', cv=10)
knn_sf_acc = cv(knn_pipeline, sf_x, sf_y, scoring='accuracy', cv=10)
svm_sf_acc = cv(svm_pipeline, sf_x, sf_y, scoring='accuracy', cv=10)
```

### All–features dataset accuracies

```python
lr_af_acc = cv(lr_pipeline, af_x, af_y, scoring='accuracy', cv=10)
nb_af_acc = cv(nb_pipeline, af_x, af_y, scoring='accuracy', cv=10)
knn_af_acc = cv(knn_pipeline, af_x, af_y, scoring='accuracy', cv=10)
svm_af_acc = cv(svm_pipeline, af_x, af_y, scoring='accuracy', cv=10)
```

### Evaluation table

```python
data = {
    'Selected Features': [lr_sf_acc.mean(), nb_sf_acc.mean(), knn_sf_acc.mean(), svm_sf_acc.mean()],
    'All Features': [lr_af_acc.mean(), nb_af_acc.mean(), knn_af_acc.mean(), svm_af_acc.mean()]
}
index_names = ['Logistic Regression', 'Naive Bayes', 'KNN', 'SVM']

table = pd.DataFrame(data, index=index_names)
table
```
